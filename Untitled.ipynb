```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report

# Assuming df is your DataFrame
selected_columns = ['basematerial', 'extr_extr01_melt01_pressure_r_value_act', 'extr_extr01_melt01_temp_r_value_act', 'extr_extr01_screw_load_r_value_act', 'extr_extr01_vacuum_r_value_act', 'extr_extr01_screw_rotspeed_r_value_act']
df_selected = df_copy[selected_columns]

# Sort the DataFrame based on 'recipe'
df_sorted = df_selected.sort_values(by='basematerial')

# Separate 'recipe' as the target variable (y)
y = df_sorted['basematerial']

# Features (X) are the other columns
X = df_sorted.drop('basematerial', axis=1)

# Interpolate missing values in the features
X_filled = X.interpolate(method='linear')
#X_filled = X.ffill()

# Check for NaN values in the features after interpolation
print("NaN values in X_filled:")
print(X_filled.isnull().sum())

# Drop any rows with remaining NaN values
X_filled = X_filled.dropna()

# Verify that there are no NaN values in the target variable
print("NaN values in y:")
print(y.isnull().sum())

# Ensure indices are aligned, handling specific timestamps
common_indices = X_filled.index.intersection(y.index)
X_filled = X_filled.loc[common_indices]
y = y.loc[common_indices]

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_filled, y, test_size=0.2, random_state=42)

# Choose a classification model (e.g., Random Forest Classifier)
model = RandomForestClassifier()
model.fit(X_train, y_train)

# Make predictions on the test set
y_pred = model.predict(X_test)

# Evaluate the classification model
accuracy = accuracy_score(y_test, y_pred)
print(f'Accuracy: {accuracy}')

# Print classification report for detailed performance metrics
print('Classification Report:')
print(classification_report(y_test, y_pred))

# Print feature importance
print('Feature Importance:')
for feature, importance in zip(X.columns, model.feature_importances_):
    print(f'{feature}: {importance}')
```

# PCA

```python
import pandas as pd
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from sklearn.impute import SimpleImputer

import networkx as nx
import matplotlib.pyplot as plt
```

```python
df_PCA = df.dropna(subset=['extr_extr01_melt01_temp_r_value_act'])
```

```python
# Assuming df is your DataFrame with 90 columns
X = df_PCA

# Handling missing values by forward fill (ffill) and then using SimpleImputer
X_filled = X.ffill()

# Standardize the data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_filled)

# Impute remaining NaN or infinite values with the mean (you can choose a different strategy)
imputer = SimpleImputer(strategy='mean')
X_scaled = imputer.fit_transform(X_scaled)

# Apply PCA to reduce dimensionality
pca = PCA(n_components=0.95)
X_pca = pca.fit_transform(X_scaled)

# Create a new DataFrame with the reduced dimensionality features
#df_pca = pd.DataFrame(data=X_pca, columns=[f'PC{i}' for i in range(1, X_pca.shape[1] + 1)])

# Create a new DataFrame with the reduced dimensionality features
df_pca = pd.DataFrame(data=X_pca, columns=df_PCA.columns[:X_pca.shape[1]])


# Optional: If you want to see the variance explained by each principal component
explained_variance_ratio = pca.explained_variance_ratio_
print("Explained Variance Ratio:", explained_variance_ratio)

# Optional: If you want to see the cumulative explained variance
cumulative_explained_variance = explained_variance_ratio.cumsum()
print("Cumulative Explained Variance:", cumulative_explained_variance)

# Now, df_pca contains the reduced dimensionality features with the original column names

# Create a directed graph from the DataFrame
graph = nx.from_pandas_adjacency(df_pca.corr(), create_using=nx.DiGraph())

# # Check if the graph is a DAG
# print(nx.is_directed_acyclic_graph(graph))  # True

# # Visualize the DAG using matplotlib
# nx.draw(graph, with_labels=True)
# plt.show()

# Check if the graph is a DAG
print(nx.is_directed_acyclic_graph(graph))  # True

# Use the spring layout for a better graph representation
pos = nx.spring_layout(graph)

# Visualize the DAG using matplotlib with adjusted parameters
plt.figure(figsize=(12, 8))
nx.draw(graph, pos=pos, with_labels=True, node_size=1000, node_color="skyblue", font_size=8, font_color="black", font_weight="bold", edge_color="gray", arrowsize=15)
plt.title("Directed Acyclic Graph (DAG) Visualization")
plt.show()
```
